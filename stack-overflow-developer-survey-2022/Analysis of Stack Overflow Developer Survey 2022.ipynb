Now we are going to review that which programming language is in use the most. We will also plot the data to the graph.

Steps to follow:
1) First split the data by delimiter ";".
2) Stack the splitted data in one column.
3) Count the frequencey of each data.
4) Plot the counts of each data in a bar chart.

```python
# Split the data into columns by delimiter.
lang_split = new_dataset['LanguageHaveWorkedWith'].str.split(';', expand=True)
lang_split.head()
```

```python
# Stack splitted data into single column and get the count of each value.
stack_data = lang_split.stack()
# Count of each splitted data.
count_programming_language = stack_data.value_counts()
count_programming_language
```

```python
# Plotting splitted count data on the graph using seaborn.
#stack_data.value_counts().plot(kind='bar', figsize=(15,8), color='Green', title='Stack Overflow Developer Survey 2022', xlabel=('Name of the Programming languages'), ylabel=('Count of the programming languages'))

fig, ax, = plt.subplots(figsize=(15,15))
counts = stack_data.value_counts().rename_axis('No. of records').reset_index(name='Name of programming language')
ax = sns.barplot(x='Name of programming language', y='No. of records', data=counts)
ax.bar_label(ax.containers[0])
ax.set(xlabel='No. of records', ylabel='Name of programming language')
ax.set(title='Popularity of Language')
```
